```python
# loading the required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as datetime
from datetime import timedelta
import os
import glob
```

```python
# importing dataset
folder_path=r"C:\Users\USER\Documents\Google Capstone\Bike_Sharing_Data"
bike_share_sheets=glob.glob(os.path.join(folder_path, '*.csv'))
dataframes=[pd.read_csv(file) for file in bike_share_sheets]
```

```python
all_sheet=pd.concat(dataframes, ignore_index=True)
```

```python
# drop colums with null values
all_sheet=all_sheet.drop(columns=['start_station_name', 'start_station_id', 'end_station_name', 'end_station_id', 'start_lat', 'start_lng', 'end_lat', 'end_lng'])
```

```python
# convert started_at and ended_at to datetime
all_sheet['started_at']=pd.to_datetime(all_sheet['started_at'])
all_sheet['ended_at']=pd.to_datetime(all_sheet['ended_at'])
```

```python
# create new column Weekday and Hour columns
all_sheet['Weekday']=all_sheet['started_at'].dt.day_name()
all_sheet['Hour']=all_sheet['started_at'].dt.hour
```

```python
# create Ride_length column
all_sheet['Ride_length']=(all_sheet['ended_at']-all_sheet['started_at'])
all_sheet['Ride_length']=(all_sheet['Ride_length'])/timedelta(minutes=1)
```

```python
all_sheet.head(5)
```

```python
Total_Rider=all_sheet['ride_id'].count()
Total_Rider
```

```python
# create Month column
all_sheet['Month']=all_sheet['started_at'].dt.strftime('%b')
all_sheet['Month']
```

```python
Month_dict={'Jan':1, 'Feb':2, 'Mar':3, 'Apr':4, 'May':5, 'Jun':6, 'Jul':7, 'Aug':8, 'Sep':9, 'Oct':10, 'Nov':11, 'Dec':12}
Month_dict
```

```python
# create Month, Year column
all_sheet['Month_Year']=all_sheet['started_at'].dt.strftime('%b-%Y')
all_sheet['Month_Year']
```

```python
# Total number of riders by type of bike
All_Ridable=all_sheet.groupby('rideable_type').agg(No_Riders=('ride_id', 'count')).reset_index()
All_Ridable
```

```python
# Total number of riders by month
Rider_Month=all_sheet.groupby('Month').agg(No_Riders=('ride_id','count')).reset_index()
Rider_Month.sort_values('Month', key=lambda x: x.apply(lambda x:Month_dict[x]))
```

```python
# Total number of riders by membership type and weekday
Member_Weekday=all_sheet.groupby(['Weekday', 'member_casual']).agg(No_Riders=('ride_id', 'count')).reset_index()
Member_Weekday
```

```python
# Total number of riders by membership type
Member_Count=all_sheet.groupby(['member_casual']). agg(No_Riders=('ride_id','count')).reset_index()
Member_Count
```

```python
((Member_Count['No_Riders']/Member_Count['No_Riders'].sum()) * 100).round(2)
```

```python
# Total number of riders per hour
Member_Hour=all_sheet.groupby(['member_casual','Hour']). agg(No_Riders=('ride_id','count')).reset_index()
Member_Hour
```

```python
# Visualization
fig, axes=plt.subplots(2,1, figsize=(10,7))
palette_color=sns.color_palette('bright')
explode=[0,0.1]
axes[0].pie(Member_Count.No_Riders, labels=Member_Count.member_casual, autopct='%1.f%%', colors=palette_color)
axes[0].set_title('REGISTERED MEMBERS VS. CASUAL MEMBERS')
sns.barplot(data=Member_Hour,x='Hour', y='No_Riders', hue='member_casual', palette='Blues')
axes[1].set_title('NO OF RIDERS PER HOUR BY MEMBERSHIP TYPE')
```

```python
# Visualization
sns.set()
fig,axes=plt.subplots(2,2, figsize=(15,8))
sns.barplot(data=Member_Weekday, x='Weekday', y='No_Riders' ,hue='member_casual', palette='Blues_r', ax=axes[0,0])
axes[0,0].set_title('NO OF RIDERS PER WEEKDAY BY MEMBERSHIP TYPE')
sns.countplot(data=all_sheet, x="Month", hue="member_casual", palette='Blues_r', ax=axes[0,1])
axes[0,1].set_title('NO OF RIDERS PER MONTH BY MEMBERSHIP TYPE')
sns.barplot(data=all_sheet, x='Weekday', y='Ride_length', hue='member_casual', palette='Blues_r', ax=axes[1,0])
axes[1,0].set_title('TIME SPENT PER WEEKDAY BY MEMBERSHIP TYPE')
sns.countplot(data=all_sheet, y='rideable_type', hue='member_casual', palette='Blues_r', ax=axes[1,1])
axes[1,1].set_title('BIKE TYPE USED BY MEMBERSHIP TYPE')
plt.subplots_adjust(wspace=0.3)
plt.subplots_adjust(hspace=0.4)
```
